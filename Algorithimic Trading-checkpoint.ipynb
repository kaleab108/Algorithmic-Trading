```python
import pandas as pd
import plotly.graph_objs as go
from plotly.subplots import make_subplots
import plotly.express as px
import yfinance as yf
```

```python
# Get Apple's stock data from Yahoo finance
stock = yf.Ticker("AAPL")
stock1 = yf.Ticker("TSLA")

data = stock.history(period="1y")
data1 = stock1.history(period="1y")



print(stock)
print(stock1)

print(data)
print(data1)

print(data.head())
print(data1.head())
```

# Now We're implementing the momentum strategy in algorithmic trading for AAPL

```python
#Calculation of Momentum for AAPL
data['momentum']=data['Close'].pct_change()

#Creating subplots to show momentum and buying/selling markers
figure=make_subplots(rows=2, cols=1)
figure.add_trace(go.Scatter(x=data.index, y=data['Close'], name='Close Price'))
figure.add_trace(go.Scatter(x=data.index, y=data['momentum'], name='Momentum', yaxis='y2'))

#Adding the buy and sell signals
figure.add_trace(go.Scatter(x=data.loc[data['momentum'] > 0].index, y=data.loc[data['momentum'] > 0] ['Close'],
                           mode='markers', name='Buy', marker=dict(color='green', symbol='triangle-up')))

figure.add_trace(go.Scatter(x=data.loc[data['momentum'] < 0].index, y=data.loc[data['momentum'] < 0]['Close'],
                           mode='markers', name='Sell', marker=dict(color='red', symbol='triangle-down')))

#Layout of the display of the stock on a graph
figure.update_layout(title='Algorthmic Trading using Momentum Strategy',
                    xaxis_title='Date',
                    yaxis_title='Price')

figure.update_yaxes(title="Momentum", secondary_y=True)
figure.show()
```
